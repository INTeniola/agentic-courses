> **Important:** Start small. Automate one repeat task first.

## Workflow Basics
A workflow is a connected sequence of steps.

| Part | Meaning | Example |
|---|---|---|
| Trigger | What starts the workflow | Form submitted |
| Action | What the system does | AI generates a response |
| Output | Where result goes | Send email/WhatsApp follow-up |

## Use Cases
- Student FAQ auto-replies
- Small business lead follow-up
- Content repurposing pipeline
- Weekly progress reminder messages


```python
workflow_example = {
    "trigger": "Google Form submission",
    "steps": [
        "Read form fields (name, question, email)",
        "Send question to AI model",
        "Format response in friendly tone",
        "Send follow-up message automatically"
    ],
    "output": "Email reply sent + record stored in spreadsheet"
}

for k, v in workflow_example.items():
    print(f"{k.upper()}: {v}\n")
```

## Live Demo Walkthrough (n8n)
### Demo: Form -> AI Reply -> Auto Follow-up

1. Create a form with name/email/question.
2. Add n8n trigger node (new form submission).
3. Pass question text to AI node.
4. Add formatting node for clear response.
5. Send output to email/WhatsApp follow-up node.

### Visual Breakdown
`Trigger` -> `Read Data` -> `AI Generate` -> `Format` -> `Send`

> **Tip:** Keep your first workflow under 5 nodes.


```python
tasks = [
    "Replying to common customer questions",
    "Turning long notes into summaries",
    "Sending reminders every Monday",
    "Generating weekly social captions"
]

print("Automation Priority Matrix")
for i, task in enumerate(tasks, start=1):
    print(f"{i}. {task}")

print("\nChoose 1 task that is repetitive and time-consuming. That is your first automation candidate.")
```
