## Main Lesson Sections
### 1) What is AI & LLMs?
Think of a Large Language Model as a highly capable intern that reads, writes, and processes language extremely fast. It operates on prediction, guessing the next most logical word.

### 2) The Rise of "Vibe Coding"
Programming is changing. Natural language is becoming the primary interface. Instead of writing code lines manually, you describe what you want in plain English ("vibe coding") and the AI generates the system.

### 3) Prompting Blueprint
To get quality results, use the **4-Step Blueprint**:
- **Role:** Who is the AI acting as?
- **Task:** What do you want it to do?
- **Context:** What constraints or background details does it have?
- **Format:** How should the response look (bullet points, email, markdown)?

### 4) Fine-Tuning Parameters: Temperature
Models have dials you can adjust. **Temperature** controls randomness:
- **Low Temperature (0.0 - 0.2):** Factual, logical, highly predictable (good for code, math, factual research).
- **High Temperature (0.7 - 1.0):** Creative, random, highly expressive (good for brainstorming, writing stories).
