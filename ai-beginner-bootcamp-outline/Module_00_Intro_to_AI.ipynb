## Main Lesson Sections
### 1) What AI Is
AI is software that helps with language, ideas, summaries, classification, and decision support.

### 2) Myth vs Reality
| Myth | Reality |
|---|---|
| AI knows everything | AI can be wrong and should be verified |
| AI will replace everyone now | AI usually changes tasks before full roles |
| AI is for programmers only | Beginners and non-technical users benefit daily |

### 3) AI in Daily Life
- Student revision planning
- CV and cover-letter improvement
- Small business content ideas
- Email writing and rewriting

### 4) AI vs Jobs
The better framing is **AI + Human**. People who adapt tools quickly gain leverage.

### 5) Tool Overview
- **ChatGPT**: general productivity and writing
- **Claude**: longer document reasoning and rewriting
- **Google AI Studio**: Gemini experimentation


## Live Demo Walkthrough — ChatGPT Prompts
Copy these prompts into your AI tool and compare results.


```python
prompt_email = """
You are a professional communication assistant.
Write a polite email to my manager asking for a one-day extension.
Context: I underestimated data cleaning, work is 70% complete, final delivery tomorrow by 3 PM.
Output: subject line + concise email body.
Tone: respectful, accountable, clear.
"""
print(prompt_email)
```

```python
prompt_plan = """
Act as a productivity coach.
Create a 3-day interview preparation plan for a customer success role.
I have 2 hours each evening.
Output format: day-by-day checklist, estimated time, confidence tip.
"""

prompt_ideas = """
You are a creative strategist.
Generate 15 social media ideas for a home-made snacks business in Nigeria.
Audience: students and young professionals.
Output: idea title, angle, and call-to-action.
"""

print(prompt_plan)
print("-" * 60)
print(prompt_ideas)
```
