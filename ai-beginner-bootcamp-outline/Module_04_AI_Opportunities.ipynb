### Positioning Formula
**Who you help + Problem you solve + Result you deliver**

Example:
"I help final-year students improve their CVs so they can apply with confidence and get more interview chances."


```python
positioning_template = "I help {audience} solve {problem} using {method} so they can achieve {result}."
example = positioning_template.format(
    audience="small business owners",
    problem="inconsistent social media content",
    method="AI-assisted weekly content planning",
    result="more visibility and faster content production"
)
print(example)
```
