## Main Lesson Sections
### 1) Required Developer Tools
- **IDE (VS Code / Jupyter):** Where developers write, run, and debug code.
- **Google AI Studio:** Google's playground to test prompts, adjust settings, and generate free API Keys.

### 2) What is an API?
An Application Programming Interface (API) is the waiter in a restaurant. You (the client code) give the waiter your request. The waiter takes it to the kitchen (Google's AI model) and brings back the food (response text).

### 3) Python Code Basics
- **Variables:** Named boxes for storing values (e.g., `api_key = "XYZ"`).
- **Functions:** Reusable actions (e.g., `send_prompt(text)`).
- **Parameters:** Tuning variables like `temperature` or `max_output_tokens` sent inside the API request.

### 4) Fine-Tuning Custom Models
When prompting or RAG is not enough, developers **fine-tune** a model. This means training a custom version of Gemini on a labeled dataset (pairs of custom inputs and desired responses) to teach it specific writing styles or rules.
