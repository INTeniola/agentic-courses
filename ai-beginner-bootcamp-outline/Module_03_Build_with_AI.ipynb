## Live Demo Option A — Simple Rule-Based Chatbot (Local)
This is not a full LLM chatbot, but it teaches app flow clearly.


```python
def simple_chatbot(message):
    msg = message.lower()
    if "price" in msg:
        return "Our starter service starts at NGN 15,000."
    if "hello" in msg or "hi" in msg:
        return "Hello! How can I help you today?"
    if "hours" in msg:
        return "We are available Monday to Saturday, 9 AM to 6 PM."
    return "Thanks for your message. A team member will follow up shortly."

tests = ["Hi", "What is your price?", "What are your hours?", "I need support"]
for t in tests:
    print(f"User: {t}")
    print(f"Bot : {simple_chatbot(t)}\n")
```

## Live Demo Option B — Tiny Sentiment Classifier (Keyword-Based)
A very beginner-friendly classification example.


```python
positive_words = {"good", "great", "excellent", "love", "amazing"}
negative_words = {"bad", "poor", "hate", "terrible", "slow"}

def classify_sentiment(text):
    words = set(text.lower().split())
    pos = len(words & positive_words)
    neg = len(words & negative_words)
    if pos > neg:
        return "Positive"
    if neg > pos:
        return "Negative"
    return "Neutral"

samples = [
    "This service is excellent and amazing",
    "The delivery was slow and terrible",
    "The product arrived"
]

for s in samples:
    print(f"{s} -> {classify_sentiment(s)}")
```
