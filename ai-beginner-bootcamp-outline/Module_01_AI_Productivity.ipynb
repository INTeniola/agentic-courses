## Main Lesson Sections
### 1) Search Grounding & Hallucinations
LLMs do not know real-time facts out of the box. **Search Grounding** connects the AI model to live search engines. This anchors the model's response to verified web search queries, stopping it from making up facts ("hallucinating").

### 2) What are Embeddings?
AI represents words, sentences, or documents as mathematical numbers (coordinates) on a map. Words with similar meanings are plotted close together. This allows AI to search, classify, and compare text instantly.

### 3) RAG (Retrieval-Augmented Generation)
RAG lets you bring specialist knowledge into your model. When you upload a PDF or file to an AI tool (like NotebookLM or Gemini), the system indexes the document using embeddings and retrieves relevant chunks to answer your questions.

### 4) Session Memory & Context Window
AI models have a "short-term memory" capacity known as the **Context Window**. If a conversation gets too long, or your files are too massive, the model starts to forget older parts of the conversation.

## 🚀 Interactive Demo: How Embeddings Work
Embeddings turn words into math. Let's calculate the "distance" between words to see how AI understands similarity! Run the code below.


```python
# 1. Run this cell to see mathematical similarity between words!
import math

# We represent words as simplified 2D arrays (embeddings)
# In reality, models use thousands of dimensions!
word_embeddings = {
    "Dog": [0.8, 0.9],   
    "Cat": [0.7, 0.8],   
    "Car": [0.1, 0.2]
}

def cosine_similarity(vec1, vec2):
    dot_product = sum(a * b for a, b in zip(vec1, vec2))
    mag1 = math.sqrt(sum(a * a for a in vec1))
    mag2 = math.sqrt(sum(b * b for b in vec2))
    return dot_product / (mag1 * mag2)

print("Calculating AI Similarity Scores:\n")

sim_dog_cat = cosine_similarity(word_embeddings["Dog"], word_embeddings["Cat"])
print(f"Similarity ('Dog' and 'Cat'): {sim_dog_cat:.2%} - Very High! (Both are pets)")

sim_dog_car = cosine_similarity(word_embeddings["Dog"], word_embeddings["Car"])
print(f"Similarity ('Dog' and 'Car'): {sim_dog_car:.2%} - Very Low! (Unrelated concepts)")
```
