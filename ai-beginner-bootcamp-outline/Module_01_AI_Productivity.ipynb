## Core Prompt Framework
Use: **Role + Task + Context + Format + Constraints**

> **Tip:** If output is weak, improve context first before trying another tool.

## Everyday Use Cases
1. Study support: summarization and quiz generation
2. Work support: polished emails, meeting notes, action items
3. Content support: post ideas and caption drafts
4. Planning support: daily and weekly schedules


```python
prompt_templates = {
    "study_summary": "Summarize this text for a beginner, then generate 5 practice questions.",
    "work_email": "Rewrite this draft email in a professional and polite tone.",
    "meeting_notes": "Convert these rough notes into clear minutes with action items.",
    "social_caption": "Create 10 Instagram captions for a small business with CTA included."
}

for name, prompt in prompt_templates.items():
    print(f"[{name}]\n{prompt}\n")
```
