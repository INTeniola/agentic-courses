## Step 3 — Your First Tensor Operations

The **tensor** is the fundamental data structure in deep learning — think of it as a super-powered NumPy array that can also run on GPUs.

```python
# ── Tensors: the building block of deep learning ──────────

# 1. Creating tensors
scalar   = torch.tensor(3.14)                        # 0-D
vector   = torch.tensor([1.0, 2.0, 3.0])             # 1-D
matrix   = torch.tensor([[1, 2], [3, 4]], dtype=torch.float32)  # 2-D
tensor3d = torch.randn(3, 4, 4)                      # 3-D (e.g. RGB patch)

print("Scalar :", scalar)
print("Vector :", vector)
print("Matrix :\n", matrix)
print("3D shape:", tensor3d.shape)

# 2. Key tensor operations
a = torch.randn(3, 3)
b = torch.randn(3, 3)
print("\nMatrix multiply shape:", (a @ b).shape)
print("Element-wise add shape:", (a + b).shape)

# 3. Moving tensors to GPU (if available)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
t_on_device = vector.to(device)
print(f"\nVector device: {t_on_device.device}")
```

## Step 4 — NumPy ↔ PyTorch Bridge

You will frequently need to convert between NumPy arrays and PyTorch tensors.

```python
# ── NumPy <-> PyTorch conversion ──────────────────────────

# NumPy → Tensor
arr = np.array([1.0, 2.0, 3.0, 4.0])
t   = torch.from_numpy(arr)
print("NumPy → Tensor:", t)

# Tensor → NumPy
t2  = torch.tensor([10.0, 20.0, 30.0])
arr2 = t2.numpy()
print("Tensor → NumPy:", arr2)

# Autograd: tensors can track gradients automatically
x = torch.tensor(2.0, requires_grad=True)
y = x ** 3          # y = x^3
y.backward()        # compute dy/dx automatically
print(f"\nFor y = x³ at x=2: dy/dx = {x.grad} (expected 12)")
```

## Step 5 — Quick Visualisation Check

```python
# ── Matplotlib test ──────────────────────────────────────
plt.figure(figsize=(10, 3))

# Plot 1: Sigmoid activation
x = np.linspace(-6, 6, 200)
sigmoid = 1 / (1 + np.exp(-x))

plt.subplot(1, 3, 1)
plt.plot(x, sigmoid, color='#4C72B0', linewidth=2)
plt.axhline(0.5, color='gray', linestyle='--', alpha=0.5)
plt.title("Sigmoid", fontsize=12)
plt.grid(alpha=0.3)

# Plot 2: ReLU activation
relu = np.maximum(0, x)
plt.subplot(1, 3, 2)
plt.plot(x, relu, color='#DD8452', linewidth=2)
plt.title("ReLU", fontsize=12)
plt.grid(alpha=0.3)

# Plot 3: Tanh activation
plt.subplot(1, 3, 3)
plt.plot(x, np.tanh(x), color='#55A868', linewidth=2)
plt.title("Tanh", fontsize=12)
plt.grid(alpha=0.3)

plt.suptitle("Activation Functions — Preview", fontsize=13, y=1.02)
plt.tight_layout()
plt.show()
print("✅ Visualisation working perfectly!")
```
